```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
sns.set_palette("viridis")

pd.set_option("display.max_columns", None)
```

```python
# Load Dataset
df = pd.read_csv("../data/raw/dataset.csv")  # adjust path if notebook is in root
df.head()
```

```python
# Basic Info
df.info()

df.describe().T
```

```python
# Check for Missing Data
df.isnull().sum().sort_values(ascending=False)
```

```python
# Drop Unnecessary Columns
df = df.drop(columns=["id"])
```

```python
# One-Hot Encoding for Categorical Variables
cat_cols = ["gender", "region", "income_level", "education_level", "daily_role", "device_type"]

df_encoded = pd.get_dummies(df, columns=cat_cols, drop_first=True)
df_encoded.head()
```

```python
# Visualize Numeric Feature Distributions
numeric_cols = [
    "age",
    "device_hours_per_day",
    "phone_unlocks",
    "notifications_per_day",
    "social_media_mins",
    "study_mins",
    "physical_activity_days",
    "sleep_hours",
    "sleep_quality",
    "anxiety_score",
    "depression_score",
    "stress_level",
    "happiness_score",
    "focus_score",
    "productivity_score",
    "digital_dependence_score"
]

df[numeric_cols].hist(figsize=(15, 12), bins=20)
plt.suptitle("Numeric Feature Distributions", fontsize=18)
plt.show()
```

```python
# Correlation Heatmap for Feature Selection
plt.figure(figsize=(14, 10))
sns.heatmap(df_encoded.corr(), cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap", fontsize=16)
plt.show()
```

```python
# Chosen target variable: "happiness_score"
target = "happiness_score"
```

```python
# Select Feature Columns
feature_cols = [
    "age",
    "device_hours_per_day",
    "phone_unlocks",
    "notifications_per_day",
    "social_media_mins",
    "study_mins",
    "physical_activity_days",
    "sleep_hours",
    "sleep_quality",
    "anxiety_score",
    "depression_score",
    "stress_level",
    "focus_score",
    "productivity_score",
    "digital_dependence_score"
]
```

```python
# Feature vs Target Plots
corr_with_target = (
    df_encoded[feature_cols]
    .corrwith(df_encoded[target])
    .sort_values(ascending=False)
    .to_frame(name=f"corr_with_{target}")
)

corr_with_target
```
